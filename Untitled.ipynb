# **CREAR DATAFRAME**

```python
import pandas as pd
import glob
pd.set_option('display.max_columns', None) #no funcion√≥
csv_files = glob.glob("DATA/*Friday*.csv") 

df_combinado = pd.DataFrame()

for csv_file in csv_files:
    df = pd.read_csv(csv_file)
    df_combinado = pd.concat([df_combinado,df])
```
