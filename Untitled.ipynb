# **CREAR DATAFRAME**

```python
import pandas as pd
import glob
pd.set_option('display.max_columns', None) #no funcion√≥
csv_files = glob.glob("DATA/*Friday*.csv") 

df_combinado = pd.DataFrame()

for csv_file in csv_files:
    df = pd.read_csv(csv_file)
    df_combinado = pd.concat([df_combinado,df])
```

# **LIMPIEZA**

```python
df_combinado.columns = df_combinado.columns.str.strip()
#df_combinado = df_combinado.rename(columns = {' Label':'Tipo'}) tiene un espacio
df_combinado

import numpy as np
from sklearn.preprocessing import LabelEncoder
import sys
le = LabelEncoder()

# codificar la columna label
encoded_array = le.fit_transform(df_combinado["Label"])

np.set_printoptions(threshold=sys.maxsize)


# el arreglo codificado cambiamos todo valor diferente 0 como 1 para denotar que es un tr√°fico an√≥malo 
for i in range(len(encoded_array)):
    if ( encoded_array[i] != 0 ):
        encoded_array[i] = 1
    
df_combinado['Label'] = encoded_array

        
#hacer una lista de las columnas de dtype float64 e int64
columnas_float64 = list(df_combinado.select_dtypes(include="float"))
columnas_int64 = list(df_combinado.select_dtypes(include="int"))

#seleccionar las columnas que est√©n dentro de esa lista y aplicarles astype para cambiar su dtype
df_combinado[columnas_float64] = df_combinado[columnas_float64].astype('Float32')
df_combinado[columnas_int64] = df_combinado[columnas_int64].astype('Int32')

#a√±adir cambiar los valores infinitos por nulos
df_combinado.replace([np.inf,-np.inf],np.nan, inplace=True)

#clean null
df_combinado.dropna(inplace=True)

print(df_combinado.isin([np.inf,-np.inf]))

df_combinado.head()
```

# **TOMAR DATOS**

```python
# ver que no todas las label sean benignas, ver records con label no benign
benigno = df_combinado.loc[df_combinado["Label"] != 'BENIGN']
benigno

```

# **INPUT & OUTPUT SETS, FITTING THE MODEL**

```python
from sklearn.tree import DecisionTreeClassifier
df_combinado = df_combinado.fillna(0)
x = df_combinado.drop(columns=['Label'])
y = df_combinado['Label']
 
model = DecisionTreeClassifier()
model.fit(x,y)
```

# **ACCURACY OF MODEL**

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
x_train, x_test, y_train, y_test =  train_test_split(x, y, test_size=0.2)

model.fit(x_train, y_train)

predictions = model.predict(x_test)

score = accuracy_score(y_test,predictions)
score
```

# **SAVING MODEL**

```python
import joblib
joblib.dump(model, 'ENIDS-model.joblib')
```
